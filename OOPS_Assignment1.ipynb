```python
#10.How can you have multiple constructors in a Python class? Explain with an example.

#In Python, you can't have multiple constructors in the same way as some other programming languages that support method overloading with different parameter lists.
#However, you can achieve similar functionality by using default values for parameters or class methods that act as alternative constructors.
#1.USING DEFALUT VALUES:
class Rectangle:
    def __init__(self, width=0, height=0):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height
rectangle_default = Rectangle()
rectangle_custom = Rectangle(width=5, height=8)
print("Default Rectangle - Area:", rectangle_default.calculate_area())
print("Custom Rectangle - Area:", rectangle_custom.calculate_area())

#2.USING CLASS METHOD:
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

    @classmethod
    def create_square(cls, side_length):
        return cls(width=side_length, height=side_length)
rectangle_custom = Rectangle(width=5, height=8)
square = Rectangle.create_square(side_length=4)
print("Custom Rectangle - Area:", rectangle_custom.calculate_area())
print("Square - Area:", square.calculate_area())


```

```python
#5.How can you access the methods and attributes of a parent class from a child class in Python? Give an example.


#In Python, you can access the methods and attributes of a parent class from a child class using the super() function.
#The super() function returns a temporary object of the superclass, which allows you to call its methods and access its attributes.
# This is commonly used within the child class's methods to invoke the behavior of the parent class.

class Parent:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Parent class: Name - {self.name}")

class Child(Parent):
    def __init__(self, name, additional_info):

        super().__init__(name)
        self.additional_info = additional_info

    def display_info(self):

        super().display_info()
        print(f"Child class: Additional Info - {self.additional_info}")

child_instance = Child(name="Omkar", additional_info="Likes programming")

child_instance.display_info()


```

```python
#6.Discuss the use of the `super()` function in Python inheritance. When and why is it used? Provide an example.

#The super() function in Python is used in the context of inheritance to access methods or attributes from the parent (or superclass) within the child (or subclass).
#It provides a way to invoke the methods of the superclass, enabling you to extend or customize the behavior defined in the parent class.
#super() is particularly useful when working with multiple inheritance, as it helps maintain a consistent and predictable method resolution order (MRO)

#Why super() is used:
#Call Superclass Methods:
#super() allows you to call methods from the superclass, even if they are overridden in the subclass. This is useful for extending or augmenting the behavior of the parent class.

#Access Superclass Attributes:
#You can use super() to access attributes defined in the parent class, facilitating the reuse of attributes or avoiding redundancy in attribute initialization.

#Maintain Method Resolution Order (MRO):
#In multiple inheritance scenarios, where a class inherits from more than one parent, super() helps maintain a consistent MRO, ensuring a predictable order in which classes are considered when looking up methods or attributes.

#Ex.,

class Animal:
    def __init__(self, species):
        self.species = species

    def speak(self):
        print("Generic animal sound")

class Dog(Animal):
    def __init__(self, species, breed):
        super().__init__(species)
        self.breed = breed

    def speak(self):
        super().speak()
        print("Woof!")
my_dog = Dog(species="Canine", breed="Golden Retriever")

print(f"Species: {my_dog.species}")
print(f"Breed: {my_dog.breed}")
my_dog.speak()
```

```python
#12.12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an example using the `abc` module.

#Abstract Base Classes (ABCs) in Python provide a way to define abstract classes and abstract methods.
# An abstract class cannot be instantiated, and it serves as a blueprint for other classes that inherit from it.
# ABCs help enforce a certain interface by requiring derived classes to implement specific methods.

#The abc module in Python is used to work with Abstract Base Classes.
# The ABC class and the abstractmethod decorator from this module are key components in creating and working with abstract classes.

from abc import ABC, abstractmethod

# Define an abstract base class (ABC) called Shape
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Implement a concrete class Circle that inherits from Shape
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Implement another concrete class Rectangle that inherits from Shape
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)
circle_instance = Circle(radius=5)
rectangle_instance = Rectangle(length=4, width=6)
print(f"Area of the Circle: {circle_instance.area():.2f}")
print(f"Perimeter of the Circle: {circle_instance.perimeter():.2f}")

print(f"\nArea of the Rectangle: {rectangle_instance.area()}")
print(f"Perimeter of the Rectangle: {rectangle_instance.perimeter()}")

```

```python
#1.Explain the concept of encapsulation in Python. What is its role in object-oriented programming?

#Encapsulation is one of the fundamental principles of object-oriented programming (OOP) and refers to the bundling of data and the methods that operate on that data into a single unit, known as a class.
#It involves restricting access to some of the object's components and hiding the internal details of how an object works.

#The key aspects of encapsulation are:
#Data Hiding:
#The internal representation of an object is hidden from the outside world. The object's attributes are often declared as private, meaning they can only be accessed and modified within the class.
#Access Control:
#Access to the internal state of an object is controlled through methods, providing a way to read or modify the data. This allows for validation, error checking, and the enforcement of business rules.
#Modularity:
#Encapsulation promotes modularity by organizing code into self-contained units (classes). Each class encapsulates its own data and behavior, making it easier to understand and maintain.
#Code Reusability:
#By exposing a well-defined interface through methods while keeping the internal details hidden, encapsulation facilitates code reuse. Other parts of the program or other programs can use the class without needing to understand its internal implementation.

class BankAccount:
    def __init__(self, account_holder, balance):
        self._account_holder = account_holder
        self._balance = balance
    def deposit(self, amount):
        self._balance += amount

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self._balance
account = BankAccount(account_holder="John Doe", balance=1000)
account.deposit(500)
account.withdraw(200)
print(f"Account balance: ${account.get_balance()}")

```

```python
#7.Discuss the use of abstract methods and classes in achieving polymorphism in Python. Provide an example using the `abc` module.

#In Python, abstract methods and classes are a way to achieve polymorphism by defining a common interface that must be implemented by subclasses.
# Abstract methods are declared in an abstract class but don't have an implementation in the abstract class itself.
# Subclasses are required to provide concrete implementations for these abstract methods.
#The abc (Abstract Base Classes) module in Python facilitates the creation of abstract classes and abstract methods.

#Here's an example using the abc module to create an abstract class Shape with an abstract method calculate_area().
#Subclasses (Circle, Square, and Triangle) are then required to provide their own implementations of the calculate_area() method:

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius**2

class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def calculate_area(self):
        return self.side_length**2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height

shapes = [Circle(radius=2), Square(side_length=3), Triangle(base=4, height=5)]

for shape in shapes:
    print(f"Area of the shape: {shape.calculate_area():.2f}")


```

```python
#14. Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking, credit card) and demonstrate polymorphism by implementing a common `withdraw()` method.


class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Insufficient funds. Withdrawal failed.")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def withdraw(self, amount):
        super().withdraw(amount)

class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance, overdraft_limit):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount <= (self.balance + self.overdraft_limit):
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Exceeded overdraft limit. Withdrawal failed.")

class CreditCardAccount(BankAccount):
    def __init__(self, account_number, balance, credit_limit):
        super().__init__(account_number, balance)
        self.credit_limit = credit_limit

    def withdraw(self, amount):
        if amount <= self.credit_limit:
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Exceeded credit limit. Withdrawal failed.")

savings_account = SavingsAccount(account_number="SA123", balance=1000, interest_rate=0.02)
checking_account = CheckingAccount(account_number="CA456", balance=2000, overdraft_limit=500)
credit_card_account = CreditCardAccount(account_number="CC789", balance=500, credit_limit=1000)

accounts = [savings_account, checking_account, credit_card_account]

for account in accounts:
    print(f"\nAccount Type: {type(account).__name__}")
    account.withdraw(200)
```

```python
#15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators like `+` and `*`.

#Operator overloading in Python refers to the ability to define or redefine the behavior of operators for user-defined objects or classes.
#It allows you to define how an operator works for instances of your own classes, providing a way to make your objects behave like built-in types.
#Operator overloading is closely related to polymorphism because it allows objects of different types to be used with the same operators, enabling a more intuitive and flexible usage.

#Ex., '+' operator:

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other_point):
        if isinstance(other_point, Point):
            return Point(self.x + other_point.x, self.y + other_point.y)
        else:
            raise ValueError("Unsupported operand type for +: Point and {}".format(type(other_point)))

point1 = Point(1, 2)
point2 = Point(3, 4)

result_point = point1 + point2
print("Resultant Point: ({}, {})".format(result_point.x, result_point.y))


#Ex., '*' operator:

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, scalar):
       return Vector(self.x * scalar, self.y * scalar)

vector = Vector(2, 3)

result_vector = vector * 2
print("Resultant Vector: ({}, {})".format(result_vector.x, result_vector.y))


#Relation to Polymorphism:
#Operator overloading is a form of polymorphism because it allows different classes to provide their own implementations of the same operator, enabling the use of common operators with different types.
#It allows objects of different types to be used with the same operator in a way that makes sense for each type, contributing to a more flexible and intuitive code.


```

```python
#17.Create a python class hierarchy for employees in  company(e.g.,manager,developer,designer)and implement polymorphism through a common 'calculate_salary()'method.

class Employee:
    def __init__(self, employee_id, name):
        self.employee_id = employee_id
        self.name = name

    def calculate_salary(self):
        raise NotImplementedError("Subclasses must implement the calculate_salary() method")

class Manager(Employee):
    def __init__(self, employee_id, name, base_salary, bonus):
        super().__init__(employee_id, name)
        self.base_salary = base_salary
        self.bonus = bonus

    def calculate_salary(self):
        return self.base_salary + self.bonus

class Developer(Employee):
    def __init__(self, employee_id, name, hourly_rate, hours_worked):
        super().__init__(employee_id, name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked

class Designer(Employee):
    def __init__(self, employee_id, name, monthly_salary):
        super().__init__(employee_id, name)
        self.monthly_salary = monthly_salary

    def calculate_salary(self):
        return self.monthly_salary

employees = [
    Manager(employee_id=1, name="John Doe", base_salary=60000, bonus=10000),
    Developer(employee_id=2, name="Jane Smith", hourly_rate=30, hours_worked=160),
    Designer(employee_id=3, name="Bob Johnson", monthly_salary=5000)
]

for employee in employees:
    print(f"{employee.name}'s salary: ${employee.calculate_salary():,.2f}")

```

```python
#6.Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and providing methods to deposit and withdraw funds.

from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self._balance = initial_balance

    @abstractmethod
    def get_account_type(self):
        pass

    def get_balance(self):
        return self._balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            return f"Deposit of ${amount} successful. New balance: ${self._balance}"
        else:
            return "Invalid deposit amount. Please enter a positive value."

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            return f"Withdrawal of ${amount} successful. New balance: ${self._balance}"
        elif amount > self._balance:
            return "Insufficient funds. Withdrawal canceled."
        else:
            return "Invalid withdrawal amount. Please enter a positive value."

class SavingsAccount(BankAccount):
    def get_account_type(self):
        return "Savings Account"

class CheckingAccount(BankAccount):
    def get_account_type(self):
        return "Checking Account"

savings_account = SavingsAccount(account_holder="John Doe", initial_balance=1000)
checking_account = CheckingAccount(account_holder="Jane Smith", initial_balance=500)

print("Savings Account Holder:", savings_account.account_holder)
print("Savings Account Type:", savings_account.get_account_type())
print("Initial Savings Account Balance:", savings_account.get_balance())
print(savings_account.deposit(200))
print(savings_account.withdraw(50))

print("\nChecking Account Holder:", checking_account.account_holder)
print("Checking Account Type:", checking_account.get_account_type())
print("Initial Checking Account Balance:", checking_account.get_balance())
print(checking_account.deposit(100))
print(checking_account.withdraw(200))

```

```python
#12.Describe the use of abstract properties in Python and how they can be employed in abstract classes.

#Abstract properties in Python are a way to declare properties in abstract base classes (ABCs) that must be implemented by concrete subclasses.
#Properties are special attributes that are accessed like attributes but have methods associated with their getting, setting, and deleting operations.
#Abstract properties provide a mechanism to define a common interface for properties that should be present in all subclasses but require individual implementations.
#The @property decorator is used to declare an abstract property.

from abc import ABC, abstractmethod

class Shape(ABC):
    @property
    @abstractmethod
    def area(self):
        pass

    @property
    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        return 3.14 * self.radius * self.radius

    @property
    def perimeter(self):
        return 2 * 3.14 * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property
    def area(self):
        return self.length * self.width

    @property
    def perimeter(self):
        return 2 * (self.length + self.width)

circle = Circle(radius=5)
rectangle = Rectangle(length=4, width=6)

print("Circle Area:", circle.area)
print("Circle Perimeter:", circle.perimeter)

print("\nRectangle Area:", rectangle.area)
print("Rectangle Perimeter:", rectangle.perimeter)

```

```python
#13.Create a python class hierarchy for employees in a company (e.g.,manages,developer,designer) and implement abstraction by defining a common 'get_salary()'method.


from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id

    @abstractmethod
    def get_salary(self):
        pass

class Manager(Employee):
    def __init__(self, name, employee_id, team_size):
        super().__init__(name, employee_id)
        self.team_size = team_size

    def get_salary(self):
        base_salary = 60000
        bonus_per_member = 2000 * self.team_size
        total_salary = base_salary + bonus_per_member
        return total_salary

class Developer(Employee):
    def __init__(self, name, employee_id, programming_language):
        super().__init__(name, employee_id)
        self.programming_language = programming_language

    def get_salary(self):
        base_salary = 50000
        bonus_for_language = 3000 if self.programming_language == "Python" else 0
        total_salary = base_salary + bonus_for_language
        return total_salary

class Designer(Employee):
    def __init__(self, name, employee_id, years_of_experience):
        super().__init__(name, employee_id)
        self.years_of_experience = years_of_experience

    def get_salary(self):
        base_salary = 55000
        bonus_per_year = 1500 * self.years_of_experience
        total_salary = base_salary + bonus_per_year
        return total_salary

manager = Manager(name="John Manager", employee_id=101, team_size=8)
developer = Developer(name="Alice Developer", employee_id=102, programming_language="Python")
designer = Designer(name="Bob Designer", employee_id=103, years_of_experience=5)

employees = [manager, developer, designer]

for employee in employees:
    print(f"{employee.name} (ID: {employee.employee_id}) - Salary: ${employee.get_salary():,.2f}")
```

```python
#19.Create a python class for a library system,implementing abstraction by defining commom methods(e.g.,'add_book','borrow_book()') in an abstract base class.

from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self, name):
        self.name = name
        self.books = []

    @abstractmethod
    def add_book(self, book_title, author):
        pass

    @abstractmethod
    def borrow_book(self, book_title):
        pass

    def display_books(self):
        if self.books:
            print(f"Books available in {self.name} library:")
            for book in self.books:
                print(f"- {book['title']} by {book['author']}")
        else:
            print(f"No books available in {self.name} library.")

class PublicLibrary(LibrarySystem):
    def add_book(self, book_title, author):
        new_book = {'title': book_title, 'author': author}
        self.books.append(new_book)
        print(f"Book '{book_title}' by {author} added to the {self.name} library.")

    def borrow_book(self, book_title):
        for book in self.books:
            if book['title'] == book_title:
                self.books.remove(book)
                print(f"Book '{book_title}' borrowed from the {self.name} library.")
                return
        print(f"Book '{book_title}' not found in the {self.name} library.")

class SchoolLibrary(LibrarySystem):
    def add_book(self, book_title, author):
        new_book = {'title': book_title, 'author': author}
        self.books.append(new_book)
        print(f"Book '{book_title}' by {author} added to the {self.name} library.")

    def borrow_book(self, book_title):
        for book in self.books:
            if book['title'] == book_title:
                self.books.remove(book)
                print(f"Book '{book_title}' borrowed from the {self.name} library.")
                return
        print(f"Book '{book_title}' not found in the {self.name} library.")

public_library = PublicLibrary(name="City Public Library")
school_library = SchoolLibrary(name="High School Library")

public_library.add_book("The Catcher in the Rye", "J.D. Salinger")
public_library.add_book("To Kill a Mockingbird", "Harper Lee")

school_library.add_book("Introduction to Python", "John Smith")
school_library.add_book("History of World War II", "Jane Doe")

public_library.display_books()
print()

school_library.display_books()
print()

public_library.borrow_book("To Kill a Mockingbird")
school_library.borrow_book("Introduction to Python")

public_library.display_books()
print()

school_library.display_books()


```

```python
#6.Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.

class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        return f"Playing: {self.title} by {self.artist}"

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)
        return f"Added {song.title} to the playlist '{self.name}'."

    def play_playlist(self):
        if not self.songs:
            return f"The playlist '{self.name}' is empty."

        playlist_info = f"Playing playlist '{self.name}':\n"
        for song in self.songs:
            playlist_info += f"- {song.title} by {song.artist}\n"

        return playlist_info

class MusicPlayer:
    def __init__(self):
        self.playlists = []

    def create_playlist(self, name):
        new_playlist = Playlist(name)
        self.playlists.append(new_playlist)
        return f"Created playlist '{name}'."

    def add_song_to_playlist(self, playlist_name, song):
        for playlist in self.playlists:
            if playlist.name == playlist_name:
                return playlist.add_song(song)
        return f"Playlist '{playlist_name}' not found."

    def play_all_playlists(self):
        if not self.playlists:
            return "No playlists available."

        all_playlists_info = "Playing all playlists:\n"
        for playlist in self.playlists:
            all_playlists_info += playlist.play_playlist() + "\n"

        return all_playlists_info

song1 = Song(title="Song 1", artist="Artist 1", duration="3:30")
song2 = Song(title="Song 2", artist="Artist 2", duration="4:15")
song3 = Song(title="Song 3", artist="Artist 3", duration="2:50")

music_player = MusicPlayer()

music_player.create_playlist("Favorites")
music_player.add_song_to_playlist("Favorites", song1)
music_player.add_song_to_playlist("Favorites", song2)

music_player.create_playlist("Party Mix")
music_player.add_song_to_playlist("Party Mix", song2)
music_player.add_song_to_playlist("Party Mix", song3)

print(music_player.play_all_playlists())

```

```python
#13.Discuss the challenges and drawbacks of composition, such as increased complexity and potential for tight coupling between objects.

#While composition is a powerful design pattern, it does come with its own set of challenges and drawbacks.
#It's essential to be aware of these issues and carefully consider them when designing systems using composition:

#1.Increased Complexity:
#As the number of components and their interactions grows, the overall complexity of the system can increase. Managing a large number of components and their relationships may become challenging.
#Debugging and understanding the flow of control within a composed system can be more complex compared to simpler designs.

#2.Potential for Tight Coupling:
#Composition can lead to tight coupling between the delegator (containing class) and the delegate (component class). Tight coupling occurs when the delegator is highly dependent on the internal details of the delegate.
#Changes to the delegate's implementation might require modifications to the delegator, leading to a lack of flexibility.

#3.Interface Changes Impact Multiple Classes:
#If the interface of a component (delegate) changes, it may impact all the classes (delegators) that use that component. This can create a ripple effect, requiring updates to multiple classes.
#Modifications to the interface of a component may necessitate changes in various parts of the system.

#4.Initialization Order Dependencies:
#The order in which components are initialized and composed can be crucial. If dependencies between components are not managed properly, it may lead to initialization order issues.
#Components may rely on each other, and initializing them in the wrong order can result in unexpected behavior.

#5.Performance Overhead:
#Depending on the implementation, composition may introduce a slight performance overhead due to the need for additional method calls and object creation.
#While this overhead is often negligible, it may become a concern in performance-critical systems.

#6.Maintenance Challenges:
#As the system evolves, adding, removing, or modifying components may require careful consideration. Maintenance becomes challenging when changes impact multiple classes.
#Refactoring a composed system can be more intricate than modifying a simpler, less composed design.


```

```python
#14.Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def get_description(self):
        return f"{self.quantity} {self.unit} of {self.name}"

class Dish:
    def __init__(self, name, description, price, ingredients):
        self.name = name
        self.description = description
        self.price = price
        self.ingredients = ingredients

    def get_dish_info(self):
        dish_info = f"{self.name} - {self.description}\n"
        dish_info += f"Price: ${self.price}\nIngredients:\n"
        for ingredient in self.ingredients:
            dish_info += f"- {ingredient.get_description()}\n"
        return dish_info

class Menu:
    def __init__(self, name, description, dishes):
        self.name = name
        self.description = description
        self.dishes = dishes

    def get_menu_info(self):
        menu_info = f"{self.name} - {self.description}\nDishes:\n"
        for dish in self.dishes:
            menu_info += f"{dish.get_dish_info()}\n"
        return menu_info

class Restaurant:
    def __init__(self, name, location, menus):
        self.name = name
        self.location = location
        self.menus = menus

    def get_restaurant_info(self):
        restaurant_info = f"{self.name} - {self.location}\nMenus:\n"
        for menu in self.menus:
            restaurant_info += f"{menu.get_menu_info()}\n"
        return restaurant_info

ingredient1 = Ingredient(name="Tomato", quantity=2, unit="pieces")
ingredient2 = Ingredient(name="Cheese", quantity=150, unit="grams")
ingredient3 = Ingredient(name="Bread", quantity=2, unit="slices")

dish1 = Dish(name="Margherita Pizza", description="Classic tomato and cheese pizza", price=12.99, ingredients=[ingredient1, ingredient2])
dish2 = Dish(name="Grilled Cheese Sandwich", description="Toasted bread with melted cheese", price=8.99, ingredients=[ingredient2, ingredient3])

menu1 = Menu(name="Lunch Menu", description="Delicious lunch options", dishes=[dish1, dish2])

restaurant = Restaurant(name="The Tasty Spot", location="123 Main Street", menus=[menu1])

print(restaurant.get_restaurant_info())

```

```python
#16.Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.


class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def attack(self):
        return f"Attacking with {self.name}. Damage: {self.damage}"

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def absorb_damage(self, damage):
        absorbed_damage = min(damage, self.defense)
        remaining_damage = max(0, damage - self.defense)
        return f"Absorbed {absorbed_damage} damage with {self.name} armor. Remaining damage: {remaining_damage}"

class Inventory:
    def __init__(self, items=None):
        self.items = items or []

    def add_item(self, item):
        self.items.append(item)
        return f"Added {item} to the inventory."

    def view_inventory(self):
        return f"Inventory: {', '.join(self.items)}"

class GameCharacter:
    def __init__(self, name, health, weapon=None, armor=None, inventory=None):
        self.name = name
        self.health = health
        self.weapon = weapon
        self.armor = armor
        self.inventory = inventory

    def equip_weapon(self, weapon):
        self.weapon = weapon
        return f"Equipped {weapon.name}."

    def equip_armor(self, armor):
        self.armor = armor
        return f"Equipped {armor.name}."

    def use_item_from_inventory(self, item):
        if self.inventory and item in self.inventory.items:
            self.inventory.items.remove(item)
            return f"Used {item} from the inventory."
        else:
            return f"{item} not found in the inventory."

    def attack(self, target):
        if self.weapon:
            return f"{self.name} attacks {target}! {self.weapon.attack()}"
        else:
            return f"{self.name} attacks {target} with bare hands."

sword = Weapon(name="Sword", damage=20)
shield = Armor(name="Shield", defense=15)
inventory = Inventory(items=["Health Potion", "Mana Elixir"])

player = GameCharacter(name="Hero", health=100, weapon=sword, armor=shield, inventory=inventory)

print(player.attack(target="Enemy"))
print(player.equip_weapon(Weapon(name="Axe", damage=25)))
print(player.attack(target="Enemy"))
print(player.equip_armor(Armor(name="Plate Mail", defense=30)))
print(player.attack(target="Enemy"))
print(player.use_item_from_inventory("Health Potion"))
print(player.inventory.view_inventory())
```

```python
#18.Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def get_description(self):
        return f"{self.name}"

class Appliance:
    def __init__(self, name):
        self.name = name

    def operate(self):
        return f"{self.name} is operational."

class Room:
    def __init__(self, name, furniture=None, appliances=None):
        self.name = name
        self.furniture = furniture or []
        self.appliances = appliances or []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)
        return f"Added {furniture.get_description()} to {self.name}."

    def add_appliance(self, appliance):
        self.appliances.append(appliance)
        return f"Added {appliance.name} to {self.name}."

    def describe_room(self):
        room_description = f"{self.name}:\n"
        if self.furniture:
            room_description += "Furniture:\n"
            for item in self.furniture:
                room_description += f"- {item.get_description()}\n"
        if self.appliances:
            room_description += "Appliances:\n"
            for appliance in self.appliances:
                room_description += f"- {appliance.name}\n"
        return room_description

class House:
    def __init__(self, rooms=None):
        self.rooms = rooms or []

    def add_room(self, room):
        self.rooms.append(room)
        return f"Added {room.name} to the house."

    def describe_house(self):
        house_description = "House Overview:\n"
        for room in self.rooms:
            house_description += room.describe_room() + "\n"
        return house_description

sofa = Furniture(name="Sofa")
bed = Furniture(name="Bed")

kitchen_lamp = Appliance(name="Kitchen Lamp")
tv = Appliance(name="Smart TV")

living_room = Room(name="Living Room", furniture=[sofa], appliances=[tv])
bedroom = Room(name="Bedroom", furniture=[bed], appliances=[kitchen_lamp])

my_house = House(rooms=[living_room, bedroom])

print(my_house.describe_house())

```

```python
#19.How can you achieve flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime?


#Achieving flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime is a key aspect of composition.
#This flexibility is essential for creating modular and adaptable systems.
#Here are several strategies and techniques to achieve dynamic replacement or modification of composed objects:

#1.Dependency Injection:
#Use dependency injection to provide dependencies (composed objects) to a class from the outside.
#This allows you to dynamically replace or modify the dependencies without changing the class itself.
#Dependency injection can be achieved through constructor injection, method injection, or property injection.

class Car:
    def __init__(self, engine):
        self.engine = engine

    def start(self):
        return self.engine.start()

class Engine:
    def start(self):
        return "Engine started."


original_engine = Engine()
my_car = Car(engine=original_engine)

new_engine = Engine()
my_car.engine = new_engine


#2.Setter Methods:
#Provide setter methods in the containing class to allow dynamic modification of composed objects.
#This enables you to change the components of an object after its creation.

class Car:
    def __init__(self, engine):
        self._engine = engine

    @property
    def engine(self):
        return self._engine

    @engine.setter
    def engine(self, new_engine):
        self._engine = new_engine

    def start(self):
        return self._engine.start()

original_engine = Engine()
my_car = Car(engine=original_engine)

new_engine = Engine()
my_car.engine = new_engine


#3.Abstract Classes and Interfaces:
#Define abstract classes or interfaces for the components, allowing multiple implementations.
#Swap the implementation dynamically by providing different concrete classes at runtime.

from abc import ABC, abstractmethod

class Engine(ABC):
    @abstractmethod
    def start(self):
        pass

class ElectricEngine(Engine):
    def start(self):
        return "Electric engine started."

class GasolineEngine(Engine):
    def start(self):
        return "Gasoline engine started."

class Car:
    def __init__(self, engine):
        self.engine = engine

    def start(self):
        return self.engine.start()

electric_engine = ElectricEngine()
gasoline_engine = GasolineEngine()

my_car = Car(engine=electric_engine)
print(my_car.start())

my_car.engine = gasoline_engine
print(my_car.start())


#4.Factory Pattern:
#Use a factory pattern to create instances of composed objects.
#The factory can produce different implementations of components based on certain conditions or configuration.

class Engine:
    def start(self):
        pass

class ElectricEngine(Engine):
    def start(self):
        return "Electric engine started."

class GasolineEngine(Engine):
    def start(self):
        return "Gasoline engine started."

class EngineFactory:
    def create_engine(self, engine_type):
        if engine_type == 'electric':
            return ElectricEngine()
        elif engine_type == 'gasoline':
            return GasolineEngine()

class Car:
    def __init__(self, engine):
        self.engine = engine

    def start(self):
        return self.engine.start()

engine_factory = EngineFactory()

electric_engine = engine_factory.create_engine('electric')
my_car = Car(engine=electric_engine)
print(my_car.start())

gasoline_engine = engine_factory.create_engine('gasoline')
my_car.engine = gasoline_engine
print(my_car.start())

```

```python
#20.Create a Python class for a social media application, using composition to represent users, posts, and comments.


class Comment:
    def __init__(self, user, text):
        self.user = user
        self.text = text

    def __str__(self):
        return f"{self.user.username}: {self.text}"

class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_comments(self):
        return "\n".join(str(comment) for comment in self.comments)

    def __str__(self):
        return f"{self.user.username} posted:\n{self.content}\nComments:\n{self.display_comments()}"

class User:
    def __init__(self, username, bio):
        self.username = username
        self.bio = bio
        self.posts = []

    def create_post(self, content):
        post = Post(user=self, content=content)
        self.posts.append(post)
        return post

    def comment_on_post(self, post, text):
        comment = Comment(user=self, text=text)
        post.add_comment(comment)
        return comment

    def display_posts(self):
        return "\n".join(str(post) for post in self.posts)

user1 = User(username="Alice", bio="Adventure seeker")
user2 = User(username="Bob", bio="Tech enthusiast")

post1 = user1.create_post(content="Exploring new places!")
user2.comment_on_post(post=post1, text="Looks amazing!")
user1.comment_on_post(post=post1, text="Thanks! It was an incredible journey.")

post2 = user2.create_post(content="Just built a cool app!")
user1.comment_on_post(post=post2, text="Impressive work!")
user2.comment_on_post(post=post2, text="Glad you like it!")

print(user1.display_posts())
print(user2.display_posts())

```
